```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np

# Parameters
k = 0.006
M = 10**13
N0 = 10**9
t_max = 1200

# Gompertz growth model function
def gompertz_growth(N0, k, M, t):
    return M * np.exp(np.log(N0 / M) * np.exp(-k * t))

# Time array
time_steps = 1000
ListOfTimes = np.linspace(0, t_max, time_steps)

# Simulate growth using arrays and loops
N = np.zeros(time_steps)

for singleTime in range(time_steps):
    N[singleTime] = gompertz_growth(N0, k, M, ListOfTimes[singleTime])

# N now contains the simulated growth values

def DisplayGraph(Values,Times,mValue):
    plt.plot(Times, Values, label=f'M={M:.0e}')
    plt.xlabel('Time')
    plt.ylabel('Number of Cells')
    plt.title('Gompertz Model of Tumor Growth')
    plt.legend()
    plt.grid(True)
    plt.show()


DisplayGraph(N,ListOfTimes,M)

```
