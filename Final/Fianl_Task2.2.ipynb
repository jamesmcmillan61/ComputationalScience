```python
# Imports reqired packages 
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Preset peramiters 
GridSize = 20
k = 0.006
N0 = 10**9
M = 11**13
h = 0.001


```

```python
# Function to simulate a signle cell's crowth 
def SignleCellsGrowth(h,k,M,InishalN):
    print("Single Cells growth")
    maxHit = False

    previousValue = InishalN
    ThisValue = 0

    t = 0

    while(maxHit == False):
        
        ThisValue = previousValue + h * k * previousValue * np.log(M / previousValue)

        if(HasItReachedSteadyState(previousValue, ThisValue) == True):
            break
        else:
            previousValue = ThisValue
            t = t + h

        #print("Another cycle")
    return True

```

```python
# Function to determine if the new value is at the steady state 
def HasItReachedSteadyState(PreviousValue, NewValue):
    target = 1.00001 * PreviousValue
    # If the previous value and the new value are very close to each other
    if NewValue <= target and NewValue >= PreviousValue:
        return True
    else:
        return False
```

```python
# Function for visulisign the grid
def VisuliseGrid(grid):

    plt.imshow(grid, cmap='hot', interpolation='nearest')
    plt.show()
    return
```

```python
# Choose a new location to move to 
def MoveLocation(Current_x, Current_y, Grid):
    # Generates a list of avalible moves, if avalible moves = 0 then retunrs the values given as a end value 

    # Generates a list of all of the positions that havent been filled - 
    UnfilledPositions = np.where(Grid == 0)
    UnfilledSet = set(zip(UnfilledPositions[0], UnfilledPositions[1]))

 #   print(UnfilledSet)
    # Calculates what cells are around it 
    CellsAround = [(Current_x - 1, Current_y - 1), (Current_x - 1, Current_y), (Current_x - 1, Current_y + 1),(Current_x, Current_y - 1), (Current_x, Current_y + 1),(Current_x + 1, Current_y - 1), (Current_x + 1, Current_y), (Current_x + 1, Current_y + 1)]

#    print(CellsAround)
    # checks to see if these cells are in the UnfilledPositions 
    ValliedCells = [cell for cell in CellsAround if cell in UnfilledSet]

#    print(ValliedCells)
    # checks if their is any, and picks one of these 
    if ValliedCells:
        print("Has values")

        # Selects one of the values from the list, does this by picking from the list using a unifrom distribution
        Selected = random.choice(ValliedCells)
        return Selected[0], Selected[1]
    else:
        print("No Values left ****")
        return Current_x, Current_y
    # returns this 


    return None
```

```python
# Creates a 2D array of a Grid - where cell is 0 its not used 
Grid = np.zeros((GridSize, GridSize))

# Calculates the inishal starting point -- assumes grid is squair 
StartingLocation = GridSize//2

Grid[StartingLocation,StartingLocation] = 1


currentPostion_X = StartingLocation
currentPostion_Y = StartingLocation

GrowthEnded = False
```

```python
# Simulating the growth
while(GrowthEnded == False):
    # Runs the growth on this cell 
    done = SignleCellsGrowth(h,k,M,N0)
    # cells growth simulated so marks as filled 
    Grid[currentPostion_X,currentPostion_Y] = 1

    # Visulise the grid with new one now added
    VisuliseGrid(Grid)

    # Calculates where to move to next 
    New_X,New_Y = MoveLocation(currentPostion_X, currentPostion_Y, Grid)

    print(New_X, New_Y)

    if(New_X == currentPostion_X and New_Y == currentPostion_Y):
        # No where for the simulation to move to so ends
        GrowthEnded = True
        break
    else:
        currentPostion_X = New_X
        currentPostion_Y = New_Y
```
