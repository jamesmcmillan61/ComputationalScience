```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calculating the steady state 
def HitStreadyState(Times, values,time_steps):
    steady_state_threshold = 0.000008 # when change is very small
    steady_state_time = None 
    for i in range(1, time_steps): 
        if abs(values[i] - values[i-1]) < steady_state_threshold: 
            steady_state_time = Times[i] 
            break

    # Output the steady state time 
    if steady_state_time: 
        print(f"Steady state reached at time: {steady_state_time}") 
    else: 
        print("Steady state not reached within the given time range.")
```

```python
# Methord for generating a graph based on the results of the gomertz model 
def DisplayGraph(Values,Times,mValue):
    plt.plot(Times, Values, label=f'M={mValue:.0e}')
    plt.xlabel('Time')
    plt.ylabel('Number of Cells')
    plt.title('Gompertz Model of Tumor Growth')
    plt.legend()
    plt.grid(True)
    plt.show()


#DisplayGraph(N,ListOfTimes,M)
#HitStreadyState(ListOfTimes,N)
```

```python
# Gompertz growth model 
def gompertz_growth(N0, k, M, t):
    return M * np.exp(np.log(N0 / M) * np.exp(-k * t))
```

```python

# Parameters
k = 0.006
#M = 10**13
N0 = 10**9
t_max = 1200
h = t_max / 1200
time_steps = int(t_max / h)

```

```python
def CalculateForM(SelectedM):
    # Generates a list of times that will be calculated at
    ListOfTimes = np.linspace(0, t_max, time_steps)

    # Simulate growth using arrays and loops
    N = np.zeros(time_steps)

    for singleTime in range(time_steps):
        N[singleTime] = gompertz_growth(N0, k, SelectedM, ListOfTimes[singleTime])


    DisplayGraph(N,ListOfTimes,SelectedM)
    HitStreadyState(ListOfTimes,N,time_steps)  


    return ListOfTimes, N
    
```

```python
print("M - 10^13")
CalculateForM(10**13)

print("---")

print("Where M is smaller ")

print("M - 10^1")
CalculateForM(10**1)

print("---")

print("M - 9^13")
CalculateForM(9**13)

print("---")


print("Where M is bigger ")

print("M - 10^13")
CalculateForM(10**18)

print("---")

print("M - 10^13")
CalculateForM(12**13)

print("---")
```
