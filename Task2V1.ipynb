```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
k = 0.006
M = 10**13
N0 = 10**9
t_max = 1200

# Gompertz growth model function
def gompertz_growth(N0, k, M, t):
    return M * np.exp(np.log(N0 / M) * np.exp(-k * t))

# Time array
time = np.linspace(0, t_max, 1000)

# Simulate growth
N = gompertz_growth(N0, k, M, time)

# Plotting the results
plt.plot(time, N, label=f'M={M:.0e}')
plt.xlabel('Time')
plt.ylabel('Number of Cells')
plt.title('Gompertz Model of Tumor Growth')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Function to determine if growth has reached a steady state
def has_steady_state(N, threshold=0.01):
    return np.all(np.abs(np.diff(N[-10:]) / N[-10:][:-1]) < threshold)

# Check for steady state at t=1200
steady_state = has_steady_state(N)
print(f'Steady state reached by t={t_max}: {steady_state}')

# Experiment with different values of M
M_values = [0.5 * 10**13, 1.5 * 10**13]
for M_new in M_values:
    N_new = gompertz_growth(N0, k, M_new, time)
    plt.plot(time, N_new, label=f'M={M_new:.0e}')

plt.xlabel('Time')
plt.ylabel('Number of Cells')
plt.title('Gompertz Model with Different Capacities')
plt.legend()
plt.grid(True)
plt.show()
```
